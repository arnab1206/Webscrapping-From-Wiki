```python
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
import numpy as np
```

```python
from pandas.io.html import read_html
page = 'https://en.wikipedia.org/wiki/Taoiseach'
```

```python
wikitables = read_html(page, attrs={"class":"wikitable"})
```

```python
print("Extracted {num} wikitables".format(num=len(wikitables)))
```

```python
df = wikitables[0][1:]
df = df.drop([0,1], axis = 1)
df2 = df[2].str.split('(', expand = True)
df3 = df2[1].str.split(')', expand = True)
df4 = df3.drop(0, axis="columns")
df5 = df4[1].str.split('for ', expand = True)
df6 = df5.drop(0, axis="columns")
df6 = df5[1].str.split('until', expand = True)
df6 = df6.drop(1, axis="columns")
df7 = df5[2].str.split('from', expand = True)
df8 = df7.drop(1, axis="columns")
pd1 = pd.concat([df, df2[0]], axis='columns')
pd2 = pd.concat([pd1, df6], axis='columns')
pd3 = pd.concat([pd2, df8], axis='columns')
pd3 = pd3.drop(2, axis="columns")
pd4 = pd3.drop(pd3.index[[9,10]])
pd5 = pd4.drop([7], axis = 1)
tof1 = pd5[3].str.split('[', expand = True)
tof2 = pd5[4].str.split('[', expand = True)
tof1 = tof1.drop(1, axis="columns")
tof2 = tof2.drop(1, axis="columns")
tof = pd.concat([tof1, tof2], axis='columns')
pd5 = pd.concat([tof, pd5], axis='columns')
pd5 = pd5.drop(4, axis="columns")
pd5 = pd5.drop(3, axis="columns")
pd6 = pd5[10].str.split('(', expand = True)
pd5 = pd.concat([pd5, pd6], axis='columns')
pd5 = pd5.drop(10, axis="columns")
pd5 = pd5.drop(1, axis="columns")
```

```python
pd5.columns = ["Start_of_office_term", "End_of_office_term", "Party", "CouncilComposition", "VicePresident", "Vice_President", "Name", "Constitutency1", "Constitutency2", "Dail"]
```

```python
pd5 = pd5.drop([1], axis = 0)
pd5 = pd5.drop(['VicePresident'], axis="columns")
pd5["Start_of_office_term"] = pd.to_datetime(pd5["Start_of_office_term"])
pd5.loc[pd5['End_of_office_term'] == "Incumbent", 'End_of_office_term'] = datetime.today()
pd5["End_of_office_term"] = pd.to_datetime(pd5["End_of_office_term"])
pd5 = pd5[['Name', 'Constitutency1', 'Constitutency2', 'Start_of_office_term', 'End_of_office_term', 'Party', 'CouncilComposition', 'Vice_President', 'Dail']]
FinalData = pd5.drop(["CouncilComposition"], axis="columns")
```

```python
FinalData
```

```python
FinalData = FinalData.reset_index()
FinalData = FinalData.drop(["index"], axis="columns")
Name = FinalData["Name"]
Name = Name.str.replace(' ','_')
```

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url2 = []
for i in Name:
    url = "https://en.wikipedia.org/wiki/{Name}".format(Name = i)
    print(url)
    url2.append(url)
```

```python
birthdate = []
for a in url2:
    response = requests.get(a)
    soup = BeautifulSoup(response.content, "html.parser")
    date = soup.find(attrs={"class" : "bday"})
    print("date : ", date.text)
    birthdate.append(date.text)
```

```python
birthdate = pd.Series(birthdate)
FinalData = pd.concat([FinalData, birthdate], axis='columns')
FinalData.columns = ["Name", "Constitutency1", "Constitutency2", "Start_of_office_term", "End_of_office_term", "Party", "Vice_President", "Dail", "Birthdate"]
DataList = FinalData.values.tolist()
```
